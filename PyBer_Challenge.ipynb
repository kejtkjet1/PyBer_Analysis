### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# Create the DataFrame by city type. This step may not be necesary for the calculation
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
```

```python
#  1. Get the total rides for each city type

ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
# Urban  count -other method
#urban_ride_count = urban_cities_df.groupby(["type"]).count()["ride_id"]
# Suburban  count
#suburban_ride_count = suburban_cities_df.groupby(["type"]).count()["ride_id"]
# Rural  count
#rural_ride_count = rural_cities_df.groupby(["type"]).count()["ride_id"]

ride_count
```

```python
# 2. Get the total drivers for each city type
driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
# Urban driver count -other method
#urban_driver_count = urban_cities_df.groupby(["type"]).count()["driver_count"]
# Suburban driver count
#suburban_driver_count = suburban_cities_df.groupby(["type"]).count()["driver_count"]
# Rural driver count
#rural_driver_count = rural_cities_df.groupby(["type"]).count()["driver_count"]
driver_count
```

```python
#  3. Get the total amount of fares for each city type
fare_count = pyber_data_df.groupby(["type"]).sum()["fare"]
fare_count
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_by_type = fare_count/ride_count
avg_fare_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_by_drivery_by_type=fare_count/driver_count
avg_fare_by_drivery_by_type
```

```python
#  6. Create a PyBer summary DataFrame. 

# Create a DataFrame
pyber_summary_df = pd.DataFrame({"Total Rides": ride_count, 
          "Total Drivers": driver_count, 
          "Total Fares": fare_count,
          "Average Fare per Ride": avg_fare_by_type, 
          "Average Fare per Driver": avg_fare_by_drivery_by_type})

# Display the data frame
pyber_summary_df


```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.

pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

# Display the data frame
pyber_summary_df



```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares for each date where the indices are the city type and date.
pyber_by_date_type_df = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
pyber_by_date_type_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_by_date_type_df=pyber_by_date_type_df.reset_index()

pyber_by_date_type_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_by_date_type_df_pivot=pyber_by_date_type_df.pivot(index="date", columns="type", values="fare")
pyber_by_date_type_df_pivot.head()
```

```python
list(pyber_by_date_type_df_pivot.columns)

```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
Pyber_jan_apr_df=pd.DataFrame(pyber_by_date_type_df_pivot.loc['2019-01-01':'2019-04-28'])

Pyber_jan_apr_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
Pyber_jan_apr_df.index = pd.to_datetime(Pyber_jan_apr_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
Pyber_jan_apr_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
Pyber_jan_apr_weekly_df=Pyber_jan_apr_df.resample("W").sum()
Pyber_jan_apr_weekly_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
Pyber_jan_apr_weekly_df.plot()

#Format
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.ylabel("Fare($USD)")
plt.title("Total Fare by City Type")

```
